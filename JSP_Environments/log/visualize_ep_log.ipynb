# Performance Comparison

```python
import numpy as np 
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

# DQN 

```python
episode_log_dqn = pd.read_csv('DQN_4204__20230427_085415_episode_log.txt', sep=',')
agent_log_dqn = pd.read_csv('DQN_4204__20230427_085415_agent_reward_log.txt', sep=',')
```

```python
perf_kpi_dqn = pd.DataFrame()
perf_kpi_dqn['alpha'] = episode_log_dqn['alpha'].values
perf_kpi_dqn['inventory'] = episode_log_dqn['inventory'].values
perf_kpi_dqn['order_waiting_time'] = episode_log_dqn['order_waiting_time'].values
perf_kpi_dqn['machines_working'] = episode_log_dqn['machines_working'].values
perf_kpi_dqn['total_reward'] = episode_log_dqn['total_reward'].values
perf_kpi_dqn['transp_working'] = episode_log_dqn['transp_working'].values
perf_kpi_dqn['processed_orders'] = episode_log_dqn['processed_orders'].values
```

```python
mean_alpha_dqn = perf_kpi_dqn['alpha'].mean()
mean_inventory_dqn = perf_kpi_dqn['inventory'].mean()
mean_order_waiting_time_dqn = perf_kpi_dqn['order_waiting_time'].mean()
mean_machines_working_dqn = perf_kpi_dqn['machines_working'].mean()
max_reward_dqn = perf_kpi_dqn['total_reward'].max()
mean_transp_util_dqn = perf_kpi_dqn['transp_working'].mean()
mean_throughput_dqn = perf_kpi_dqn['processed_orders'].mean()
print(mean_alpha_dqn, mean_inventory_dqn, mean_order_waiting_time_dqn, mean_machines_working_dqn, 
      max_reward_dqn, mean_transp_util_dqn, mean_throughput_dqn)
```

```python
perf_kpi_dqn.plot.line(y='alpha')
```

```python
perf_kpi_dqn.plot.line(y='inventory')
```

```python
perf_kpi_dqn.plot.line(y='order_waiting_time')
```

```python
perf_kpi_dqn.plot.line(y='machines_working')
```

# Comparison

```python
# Daten fr das Balkendiagramm
labels = ['PPO', 'FIFO', 'NJF', 'EMPTY']
values = [mean_alpha_ppo,mean_alpha_fifo,mean_alpha_njf,mean_alpha_empty]

# Erstellt ein Balkendiagramm
fig, ax = plt.subplots()
rects = ax.bar(labels, values)

# Beschriftungen hinzufgen
plt.title('Comparison of goodness of production control')
plt.xlabel('Model')
plt.ylabel('Mean Alpha')

for rect in rects:
    height = rect.get_height()
    ax.annotate(f'{height:.4f}',
                xy=(rect.get_x() + rect.get_width() / 2, height),
                xytext=(0, 3),
                textcoords='offset points',
                ha='center', va='bottom')

# Anzeigen des Diagramms
plt.show()
```
